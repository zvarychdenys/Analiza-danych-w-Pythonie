# Rozwiazywanie kwestii brakujacych wartosci
___

Moge rozwiazywac ich kwestie na kilka sposobow:
* usuwanie rekordow zawierajacych brakujace wartosci
* reczne uzupelnianie brakujacych wartosci
* uzypelnianie brakujacych wartosciami tendencji centralnej czyli srednia(w cechach numerycznych ), mediana (w cechach porzadkowych) czy dominanta(najszczesciej potwarzajaca wartosc w cechach kategorialnych)
* uzupelnianie najbardziej prawdopodobna wartoscia przy uzyciu ML , takich jak regresja, drzewa decyzyjne , algorytm KKN (K - nearest neighbours)


dla usuwanie brakujacych wartosci uzywam zwykle funkcji `dropna()`, przyjmuje arugment jeden `how`, ktory z kolej przyjmuje wartosci `all`, `any`
* `how = 'all'` - usuwa wzystkie rzedy zawierajace NaN lub brakujace
* `how = 'any'` - usuwa okreslone rzedy zawierajace NaN lub brakujace

#### Uzupelnenia brakujacej wartosci
___
`fillna()`- moge uzupelnic brakujace wartosci, przyjmnuje jedna wartosc , ktora bedzie wstawiana w pustych pozycjach zamiast NaN , ewentualnie srednia, mediana, dominanta

```python
# w kolumnie 'age' zamiast wartosci NaN wstawiam srednia wieku pracownikow
data['age'] = data.age.fillna(data.age.mean())
data['age']
```

```python
#w kolumnie 'income' zamiast NaN wstawiam mediane zarobkow pracownikow 
data['income'] = data.income.fillna(data.income.median())
data
```

```python
# kolumnie 'gender' zamiast Nan wstawiam wartosc ktora najszczej wystepuje (dominante)
data['gender'] = data.gender.fillna(data.gender.mode()[0])
data
```

### Rozwiazywanie kwestii elementow odstajacych 
___
elemetny odstajace dzielimy na **jednowymiarowe i wielowymiarowe**


```python
import pandas as pd

data = pd.read_csv("data/csv/employee.csv")
data
```

##### Uywaninie elementow odstajacyh  za pomoca sredniej i odchylenia standarowego

```python
#obliczam gorna granice  = srednia + 3 * odchyl. standr.
upper_limit = data['performance_score'].mean() + 3 * data['performance_score'].std()

#obliczam dolna granice  = srednia - 3 * odchyl. standr.
lower_limit = data['performance_score'].mean() - 3 * data['performance_score'].std()
print(f'Dolny limit: {lower_limit}')
print(f'Gorny limit: {upper_limit}')

data = data[(data['performance_score'] < upper_limit) & (data['performance_score'] > lower_limit)]
data
```

##### Usywanie elementow odstajacyh za pomoca wartosci percentylowych
Jako dolna granice wzialem .01 percentyl , natomiast gorna granice wzilem .99 percentyl. To pozwala wykluczac odstajace elementy w DataFrame

```python
import pandas as pd
data = pd.read_csv('data/csv/employee.csv')
data
```

```python
upper_limit = data['performance_score'].quantile(.99)
lower_limit = data['performance_score'].quantile(.01)
print(f'Dolny limit: {lower_limit}')
print(f'Gorny limit: {upper_limit}')

data = data[(data['performance_score'] < upper_limit) & (data['performance_score'] > lower_limit)]
data
```

### Techniki kodowania cech
___

Modele Ml, sa modelami matematycznymi, dla tego , dla swoich olbiczen potrzebuja wartosci stale, lub zmiennoprzecinkowe. Dla tego musimy przekszalcac cechy kategorialne w numeryczne. Wydajnosc tekiego ML , zalezy od wybranej techniki kodowania.

### Kodownanie  ''goracojedynkowe" (ang. onehotencoder)
___
wartosci kategorialne zostana zatapione wartosciami binarnymi czyli *1* i *0*

```python
# Oczytuje dane
data=pd.read_csv('data/csv/employee.csv')

# Kodowanie 'goracojedynkowe'
encoded_data = pd.get_dummies(data['gender'])

# Laczy zakodowane dane z pierwotna ramka danych
data = data.join(encoded_data)
data.head()
```

##### W powyzszym listingu widze ze zostaly dodane dwie kolumny 'F' i 'M', wartosci w ktorej sa jednyki - co oznacza Truem i zera - False

##### Moge uzyskac takie same rezultaty za pomoca klasy `OneHotEncoder` z biblioteki  `scikit-learn`

```python
# imporuje koder "goracojedynkowy"
from sklearn.preprocessing import OneHotEncoder

#tworze obiekt kodera "goracojedynkowy"
onehotencoder = OneHotEncoder()

#wstawiam w kolumne 'gender' zamiat NaN , wartosc dominantna
data['gender'] = data.gender.fillna(data.gender.mode()[0])

#dopasowuje i przekszalcam kolumne gender
onehotencoder.fit_transform(data[['gender']]).toarray()
```

### Kodowanie etykietowe:

### Metody skalowanie cech
___
Porownywanie cech mieszczace sie o roznych skalach, np wiek opisujemy od 0 do 100, a wynagrodzenie od 0 do 100000 w danej walucie. Cechy w ktorych wielkosci sa bardzo duze , beda mialy w ML wieksza wage,dla tego mozemy taki problem rozwiazac za pomoca **skalowaniu cech lub normalizowaniu cech.**

Metody skalowanie cech:
> * Skalowanie standordowe lub standaryzacja - obliczane za pomoca std i srednej 
> * Skalowanie min-max
> * Skalowanie odporne

```python
#importuje klase StandardScaler
from sklearn.preprocessing import StandardScaler

#inicjalizuje klase StandardScaler
scaler = StandardScaler()

#skaluje dane
scaler.fit(data['performance_score'].values.reshape(-1,1)) 
data['performance_std_scaler']=scaler.transform(data['performance_score'].values.reshape(-1,1))
data.head()
```

### Przeksztalcenia cech
___
za pomoca **dyskretyzacji** mozemy przeksztalcac kolumne. Na przyklad przepisywanie zamiast wieku pracownikow , etykiety np. mlody, dorosly, stary

```python
data = pd.read_csv("data/csv/employee.csv")

def performance_grades(score):
    if score >= 700:
        return 'A'
    if score < 700 and score >= 500:
        return 'B'
    else:
        return 'C'
    
data['performance_grade'] = data["performance_score"].apply(performance_grades)
data
```
