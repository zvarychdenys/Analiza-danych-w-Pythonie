# Wizualizacja danych
-----
Python zawiera wiele roznych bibliotek dla wizualizacji danych np. matplotlib, seaborn , Bokeh. W tym rozdziale skoncentuje sie najpierw na podstawowej bibliotece dla wizualizacji danych **matplotlib**, potem przejde do **seaborn** , i na koniec do **Bokeh**

# Wykres liniowy
* `plt.plot()` - funckja sluzy do utworzenia bardzo prostego wykresy.

```python
import numpy as np
import matplotlib.pyplot as plt

#Tworze dane
a = np.linspace(0,20)

#Rysuje wykres
plt.plot(a, a+5, label = 'linear')

#Wyswietla wykres
plt.show()
```

### Akcesoria wykresu
* `plt.title()` - dodaje tytul do wykresu
* `plt.xlabel()` - dodaje etykieta dla osi x
* `plt.ylabel()` - dodaje etykieta dla osi y

```python
x = [1,3,5,7,9,11]
y = [10,25,35,33,41,59]

plt.plot(x,y, label = 'Zestaw 1.', color = 'blue')

x = [2,4,6,8,10,12]
y = [15,29,32,33,38,55]

plt.plot(x, y, label = 'Zestaw 2.', color = 'green')

plt.xlabel('Etykieta X', color = 'orange')
plt.ylabel('Etykieta Y', color = 'orange')

plt.title('Wykres wielolinowy w Pythonie', color = 'yellow')
plt.legend()

plt.show()
```

# Wykres punktowy
----
* `plt.scatter()`- funkcja za pomoca ktorej moge otworzyc wykres punktowy

>Atrybuty: 
>* `alpha=0.3` - przezroczystosc [0,1]
>* `marker='*'` - ksztalt punktow
>* `c = 'red'` - color

```python
import matplotlib.pyplot as plt

x = [1,3,5,7,9,11]
y = [10,25,35,33,41,59]

# alpha przezroczystosc [0,1], marker - ksztalt punktow, c - color.
plt.scatter(x,y, alpha=0.3, marker='*',c = 'red')
plt.xlabel('Etykieta x', c = 'orange')
plt.ylabel('Etykieta y', c = 'orange')
plt.title('Wykres punktowy', c = 'orange')
plt.show()
```

# Wykres kolowy
-----
* `pie()` - umozliwia storzenia wykresy kolowego

```python
subjects = ['Matematyka','Informatyka', 'Angielski', 'WF']
scores = [85,62,57,92]

#Rysuje wykres kolowy
plt.pie(scores,
        labels = subjects,
        colors = ['r','g','b','y'],
        shadow = True,
        startangle= 90,
        explode = (0,0.2,0,0),
        autopct = '%1.2f%%'
       )
plt.title('Skutecznosc uczniow')
```
