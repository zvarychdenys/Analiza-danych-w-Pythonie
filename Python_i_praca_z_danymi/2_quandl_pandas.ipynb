* **min()** - metoda zwraca min znaczenia kazdej z kolumn

```python
# biblioteka pandas umozliwia zlaczenia wielu obiektow DataFrame za pomoca funkcji merge()
dest = pd.read_csv('/book/dest.csv')
tips = pd.read_csv('/book/tips.csv')
tips.head()
```

```python
dest.head()
```

##### Złączenie wewnętrzne (ang. inner join) 
**-jest rownowazne wyznaczeniu czesci wspolnej zbiorow, sluzy jedynie do wybierania wspolnych rekordow**

```python
df_inner = pd.merge(dest,tips, on = 'EmpNr', how = 'inner')
df_inner.head()
```

##### Pełne złączenie zewnętrzne (ang. full outer join)¶
**-jest rownowazne sumie zbiorow** , wypisuje wszystko , rozbieznosci uzupelnione __NaN__

```python
df_outer = pd.merge(dest,tips , on = 'EmpNr', how = 'outer')
df_outer.head()
```

##### Prawostronne złączenie zewnętrzne (ang. right outer join)¶
**-wybierane wszytkie rekordy z prawej ramki danych**

```python
df_right = pd.merge(dest,tips, on = 'EmpNr' , how = 'right')
df_right.head()
```

##### Lewostrone złączenie zewnętrzne (ang. left outer join)¶
**-wybierane wszytkie rekordy z lewej ramki danych**

```python
df_left = pd.merge(dest,tips, on = 'EmpNr', how = 'left')
df_left.head()
```

##### Usywanie brakujacych wartosci 
funkcja `dropna()` - usywanie NaN z ramki danych

```python
df.info()
```

```python
#Zmiejszylem liczbe obserwacji z 202 do 118
df.dropna(inplace = True) 
df.info()
```

##### Uzupelnianie brakujacych watosci

# Tworzenie tabel przestawnych **(ang.pivot table)** 
tabela przestawna jest tabel podsumowujaca. Jest uzywana dla podsumowania swoich rezultatow

```python
purchase = pd.read_csv('/book/purchase.csv')
purchase
```

```python
pd.pivot_table(purchase, values= 'Number',index = ['Weather'] , columns = ['Food'], aggfunc = np.sum)
```

# Rozwiazywanie kwestii dat

 1.`data_range()` - funckja umozliwiajaca okreslenie zakresow dat

atrybuty :
                `freq = 'B'` - dni robocze , `W` - czestotliwosc tygodniowa, `D` - czestotliwosc dniowa `H` - czestotliwosc godzniowa, `M` - czestotliwosc minutowa, `S` - czestotliwosc sekundowa.
                
2. `to_datetime()` - przeksztalca znacznik czasowy w formie lancucha znakow w date

```python
#zakres dat od 1 stycznia 2000r i rozciagajacy sie na 45 dni:
pd.date_range('01-01-2000', periods = 45, freq = 'D')
```
