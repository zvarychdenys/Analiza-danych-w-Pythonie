### Wykres liniowy
___
`plt.plot()` - sluzy do tworzenia wykresu liniowego

```python
x = [1,3,5,7,9,11]
y = [10,25,35,33,41,59]

#tworze pierwszy wykres liniowy
plt.plot(x,y, label = 'Zestaw 1.', color = 'blue')

x1 = [2,4,6,8,10,12]
y1 = [15,29,32,33,38,55]

#drugi wykres liniowy
plt.plot(x1,y1, label = 'Zestw 2.', color = 'red', alpha = 0.5)

# Tworze etykiety dla x i dla y
plt.xlabel('Etykieta dla X:', c = 'green')
plt.ylabel('Etykieta dla Y:', c = 'green')

#Wyswietlam legende
plt.legend()

#Tworze tytyl wykresu
plt.title('Wykres wieloliniowy w Pythonie', c = 'green')
```

### Wykres kolowy

> **Atrybuty**:
> ____

### Wykres bablekowy 
___
Wykorzyuje wykres punktowy dla utworzenia wykresu bablekowego
* `s = scaled_gdp_per_capita`- rozmiar babelek
* `plt.figure(figsize=(9,9))`- rozmiar rysunku
* `plt.xticks(rotation = 45)` - obracanie etykiety X na 45

```python
# Tworzy dane
countries = ['Qatar','Luxembourg','Singapore','Brunei','Ireland','Norway','UAE','Kuwait']
populations = [2781682, 604245,5757499,428963,4818690,5337962,9630959,4137312]
gdp_per_capita = [130475, 106705, 100345, 79530, 78785, 74356,69382, 67000]

#wyznaczam rozmiar rysunku 
plt.figure(figsize=(9,9))

#skaluje dochod PKB per capita tak,zeby tworzyc babelki
scaled_gdp_per_capita = np.divide(gdp_per_capita,80)
color = np.random.rand(8)

#Rysuje wykres punktowy
plt.scatter(countries,populations,s = scaled_gdp_per_capita , c = color)

#Dodaje etykiety
plt.xlabel('Panstwa', color = 'red')
plt.ylabel('Populacja', color = 'green')
plt.title('Wykres babelkowy')

# W celu zwiekszenia czytelnosci obracam etykiete osi X
plt.xticks(rotation = 45)
```

### Tworzenie wykresow za pomoca biblioteki pandas
___
`pd.plot()` - pozwala tworzyc wykresy bezposrednio na podstawie obiektow DataFrame
> Do tworzenia wykresow sluza nastepujace parementry metody plot():
>___
> 1. `kind = ` - okresla typ wykresu:
> * line (liniowy)
> * bar (kolumnowy)
> * barh (kolumnowy poziomy)
> * hist (histogram)
> * box (pudelkowy)
> * kde (wykres gestosci)
> * pie (kolowy)
> * area (obszarowy)
> * scatter (punktowy)
> 2. `figsize` - definiuje rozmiar obrazu jako krotke (szerokosc, wysokosc)
> 3. `title`- wyznacza tytul wykresu
> 4. `grid` - rysuje linie siatki; przyjmuje wartosci logiczne
> 5. `legend` - definuje legende
> 6. `xticks`- okresla sekwencje jednostek w osi x
> 7. `yticks`- okresla sekwencje jednostek w osi y

### Wykres lm
___
Wykres lm jest wykresem punktowym, do ktorego zostaje dopasowany model regresji. Pozwala najlatwiej dostrzec relacje pomiedzy dwiema zmiennymi. Na takich wykresach jest rysowany wspolny rozklad prawdopodobienstwa dwoch zmiennych.

```python
#importuje wymagane bibloteki
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#tworze ramke danych
df = pd.DataFrame({
                'x':[1,3,5,7,9,11],
                'y':[10,25,35,33,41,59]
})

#tworze wykres lm za pomoca biblioteki seaborn
sns.lmplot(x = 'x', y = 'y', data = df)
```

##### Domyslnie funkcja `sns.lmplot()` dopasowuje linie regresji. Mozemy nia usunac za pomoca `fit_reg= False`

```python
sns.lmplot(x = 'x', y = 'y', data = df, fit_reg = False)
```

```python
# Wczytuje dane csv
df  = pd.read_csv("data/HR_comma_sep.csv")
df.columns
```

```python
sns.lmplot(
           x = 'satisfaction_level', 
           y = 'last_evaluation', 
           data = df, 
           fit_reg = False, 
           hue = 'left' # wyznacza odcien , zmienna left oresla czy pracownik zwolnil sie z firmy
)
```

### Tworzenie prostego wykresu
___
* `output_notebook()`- sprawie ze wykres bedzie wyswietlany w aplikacji Jupyter Notebook
* `figure`- stanowi jeden z glownych elementow rysowania wykresow, za niego pomoca okreslam tytul, rozmiar, etykiety, siatke i styl wykresu.
* `show(fig)`- funckja wyswietla wynik juz po dodaniu wszystkich cech i danych do wykresu

```python
# imporuje wymagane moduly
from bokeh.plotting import figure 
from bokeh.plotting import output_notebook
from bokeh.plotting import show 

#tworze dane
x = [1,3,5,7,9,11]
y = [10,25,35,33,41,59]

#dane wynikowe sa umieszczane w notatniku
output_notebook()

fig = figure(plot_width = 500, plot_height = 350)
fig.circle(x,y, size = 10, color = 'red', alpha = 0.7)

show(fig)
```

# Szablony
____
biblioteka Bokeh zawiera szablony do organizowania wykresow i widzetow. Szablony te zawieraja wiele wykresow w jednym panelu. 
> Rodzaje szablonow:
> * **rzedowy** - ogranizuje wszystkie wykresow w rzedzie (poziomo)
> * **kolumnowy** - ogranizuje wszystkie wykresow w kolumnie (pionowo)
> * **zagniezdzony** - stanowi polaczenia szablownow rzedowego i kolumnowego
> * **siatkowy** - tworzy siatke pozwalajaca dowolnie organizowac wykresy

### Szablony rzedowe i kolumnowe
___

```python
# importuje moduly
from bokeh.plotting import figure, output_notebook, show
from bokeh.layouts import row, column

#imporuje przykladowy zestaw danych Iris w bibliotece Bokeh jako dataframe
from bokeh.sampledata.iris import flowers as df #

# umozilwiam wyswitlania rysunku w notatniku
output_notebook()

#tworze trzy rysynki
fig1 = figure(plot_width = 300, plot_height = 300)
fig2 = figure(plot_width = 300, plot_height = 300)
fig3 = figure(plot_width = 300, plot_height = 300)

# Tworze wykresy punktowy
fig1.circle(
            df['petal_length'],
            df['sepal_length'],
            size = 8,
            color = 'green',
            alpha = 0.5
)

fig2.circle(
        df['petal_length'],
        df['sepal_width'],
        size = 8,
        color  = 'blue',
        alpha = 0.5
)

fig3.circle(
            df['petal_length'],
            df['petal_width'],
            size = 8,
            color = 'red',
            alpha = 0.5
)

#tworze szablon rzedowy
row_layout = row(fig1,fig2,fig3) # wykresy buduja sie piziomo ---

#tworze szablon kolumnowy
col_layout = column(fig1,fig2,fig3) # wykresy buduja sie pionowo I

show(row_layout)
```

### Szablon zagniezdzony za pomoca szablona rzedowego i kolumnowego
___

```python
output_notebook()

#tworzenia szablony zagniezdzonego
hasted_layout  = row(fig1, column(fig2,fig3))

show(hasted_layout)
```

### Oddzialywania , Najezdzanie kursorem
---

```python
#importuje potrzebne moduly
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import CategoricalColorMapper

#importuje przykladowy zestaw danych biblioteki Bokeh jak obiekt DataFremu
from bokeh.sampledata.iris import flowers as df

output_notebook
fig = figure(
            plot_width = 500,
            plot_height = 350,
            title = 'Zaleznosc dlugosci platka od szerokosci platka',
            x_axis_label = 'petal_length',
            y_axis_label = 'petal_width'
)

#Tworze wykres punktowy
for specie, color in zip(['setosa','virginica','versicolor'],['blue','green','red']):
    data = df[df.species == specie]
    fig.circle(
               'petal_length', 
               'petal_width',
                size = 9, 
                alpha = 0.5, 
                color = color, 
                legend_label = specie,
                source = data
)
muted_color = color, muted_alpha = 0,2
    
    
#Wyznaczam pozycje legend i strategia klikniecia
fig.legend.location = 'top_left'
fig.legend.click_policy = 'mute'

show(fig)
```

# Panele zakladek
za pomoca podpakietu `bokeh.models.widgets` importuje `Tabs, Panel`, umozliwia tworzyc wiele wykresow i szablonow w jedynym oknie.

```python
# impotuje wymagane moduly
from bokeh.plotting import figure, output_notebook, show
from bokeh.models.widgets import Tabs, Panel

# importuje dane iris
from bokeh.sampledata.iris import flowers as df

output_notebook()

#wystawiam rozmiar rysynku
fig1 = figure(plot_width = 300, plot_height = 300)
fig2 = figure(plot_width = 300, plot_height = 300)

#tworze dane fig1, fig2
fig1.circle(
            df['petal_length'],
            df['sepal_length'],
            size = 8,
            color = 'green',
            alpha = 0.5,
)

fig2.circle(
            df['petal_length'],
            df['sepal_width'],
            size = 8,
            alpha = 0.5,
            color = 'red'
)

#tworze panele
tab1 = Panel(child = fig1, title = 'Zaklada 1')
tab2 = Panel(child = fig2, title = 'Zakladka 2')

#tworze zakladke za pomoca tabeli
tab_layout = Tabs(tabs=[tab1,tab2])

show(tab_layout)
```
