##### `edgecolor` - color granic, `width` - szerokosci histogramy


```python
from collections import Counter
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

histogram = Counter(min(grade // 10 * 10,90) for grade in grades)

plt.bar([x+5 for x in histogram.keys()], histogram.values(),width=10, edgecolor = (0,0,0))
plt.axis([-10,105,0,4.5]) # definuje zakres osi x od -10 do 105, i zakres osi y od 0 do -4.5

plt.xticks([i*10 for i in range(11)]) # umieszczenie etykiet osi x w puntkach od 0 do 100
```

```python
variance     = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_erorr = [x+y for x,y in zip(variance,bias_squared)]
xs = [i for i,_ in  enumerate(variance)]
xs

plt.plot(xs, variance, 'b--' , label = 'variance')
plt.plot(xs,bias_squared, 'r-', label = "bias squared")
plt.plot(xs, total_erorr, 'g:', label = "blad calkowity")

plt.legend()
```
